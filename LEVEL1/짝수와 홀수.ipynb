## 문제
정수 num이 짝수일 경우 "Even"을 반환하고 홀수인 경우 "Odd"를 반환하는 함수, solution을 완성해주세요

```python
num =45

if num%2 ==0 :
  answer = "Even"

else:
  answer = "Odd"
```

```python
def solution(num):
    if num%2 ==0 :
        answer = "Even"

    else:
        answer = "Odd"
    return answer
```
