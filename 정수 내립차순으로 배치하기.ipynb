## 문제
함수 solution은 정수 n을 매개변수로 입력받습니다. n의 각 자릿수를 큰것부터 작은 순으로 정렬한 새로운 정수를 리턴해주세요. 예를들어 n이 118372면 873211을 리턴하면 됩니다.

## 제한사항
- n은 1이상 8000000000 이하인 자연수입니다.


```python
n=118372
new_n=str(n)
new_list = list(new_n)
print(new_list)

new_list.sort(reverse=True)
print(new_list)

answer = ""
for i in new_list:
  answer+=i
print(int(answer))
```

```python
def solution(n):  
    new_n=str(n)
    new_list = list(new_n)


    new_list.sort(reverse=True)


    answer = ""
    for i in new_list:
        answer+=i


    return int(answer)
```
